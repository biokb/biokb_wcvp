Data Models for SQL Database

```python
from sqlalchemy import Engine, create_engine
from sqlalchemy.orm import sessionmaker, Session
```

```python
from sqlalchemy import (
    Column, String, Integer, Boolean, Float, ForeignKey, Text, create_engine, JSON
)
from sqlalchemy.orm import relationship, DeclarativeBase, mapped_column
from sqlalchemy.orm import Mapped
```

```python
# Define the base class for our models
class Base(DeclarativeBase):
    pass

class Taxon(Base):
    __tablename__ = "taxon"
    
    taxonid: Mapped[int] = mapped_column(Integer, primary_key=True)
    family_id: Mapped[int] = mapped_column(Integer, ForeignKey("family.id"))
    genus_id: Mapped[int] = mapped_column(Integer, ForeignKey("genus.id"))
    taxonomicstatus_id: Mapped[int] = mapped_column(Integer, ForeignKey("taxonomic_status.id"))
    scientificnameauthorship_id: Mapped[int] = mapped_column(Integer, ForeignKey("scientific_name_authorship.id"))
    scientificname: Mapped[str] = mapped_column(String(255))
    taxonrank: Mapped[str] = mapped_column(String(50))
    references: Mapped[str] = mapped_column(String(255), nullable=True)

    # Relationships
    family: Mapped["Family"] = relationship("Family", back_populates="taxons")
    genus: Mapped["Genus"] = relationship("Genus", back_populates="taxons")
    taxonomic_status: Mapped["TaxonomicStatus"] = relationship("TaxonomicStatus")
    scientific_name_authorship: Mapped["ScientificNameAuthorship"] = relationship("ScientificNameAuthorship")
    
    # Many-to-many relationship with References
    references: Mapped[list["Reference"]] = relationship("Reference", secondary="taxon_reference")

    # Relationship with DynamicProperties
    dynamic_properties: Mapped[list["DynamicProperty"]] = relationship("DynamicProperty", back_populates="taxon")

    def __repr__(self):
        return f"<Taxon(taxonid={self.taxonid}, scientificname={self.scientificname})>"


# Family Model
class Family(Base):
    __tablename__ = "family"
    
    id: Mapped[int] = mapped_column(Integer, primary_key=True)
    name: Mapped[str] = mapped_column(String(255), unique=True)

    # Relationship with Taxon
    taxons: Mapped[list["Taxon"]] = relationship("Taxon", back_populates="family", cascade="all, delete-orphan")
    
    def __repr__(self):
        return f"<Family(name={self.name})>"

# Genus Model
class Genus(Base):
    __tablename__ = "genus"
    
    id: Mapped[int] = mapped_column(Integer, primary_key=True)
    name: Mapped[str] = mapped_column(String(255), unique=True)

    # Relationship with Taxon
    taxons: Mapped[list["Taxon"]] = relationship("Taxon", back_populates="genus", cascade="all, delete-orphan")
    
    def __repr__(self):
        return f"<Genus(name={self.name})>"

class TaxonomicStatus(Base):
    __tablename__ = "taxonomic_status"
    
    id: Mapped[int] = mapped_column(Integer, primary_key=True)
    status: Mapped[str] = mapped_column(String(100), unique=True)

    def __repr__(self):
        return f"<TaxonomicStatus(status={self.status})>"

class Reference(Base):
    __tablename__ = "reference"
    
    id: Mapped[int] = mapped_column(Integer, primary_key=True)
    citation: Mapped[str] = mapped_column(String(255))
    url: Mapped[str] = mapped_column(String(255), nullable=True)
    
    # Relationship with Taxon
    taxons: Mapped[list["Taxon"]] = relationship("Taxon", secondary="taxon_reference")

    def __repr__(self):
        return f"<Reference(citation={self.citation}, url={self.url})>"

class TaxonReference(Base):
    __tablename__ = "taxon_reference"
    
    taxon_id: Mapped[int] = mapped_column(Integer, ForeignKey("taxon.taxonid"), primary_key=True)
    reference_id: Mapped[int] = mapped_column(Integer, ForeignKey("reference.id"), primary_key=True)

class ScientificNameAuthorship(Base):
    __tablename__ = "scientific_name_authorship"
    
    id: Mapped[int] = mapped_column(Integer, primary_key=True)
    authorship: Mapped[str] = mapped_column(String(255))

    # Relationship with Taxon
    taxons: Mapped[list["Taxon"]] = relationship("Taxon", back_populates="scientific_name_authorship")

    def __repr__(self):
        return f"<ScientificNameAuthorship(authorship={self.authorship})>"

class DynamicProperty(Base):
    __tablename__ = "dynamic_property"
    
    id: Mapped[int] = mapped_column(Integer, primary_key=True)
    key: Mapped[str] = mapped_column(String(100))
    value: Mapped[str] = mapped_column(String(255))
    taxon_id: Mapped[int] = mapped_column(Integer, ForeignKey("taxon.taxonid"))

    # Relationship with Taxon
    taxon: Mapped["Taxon"] = relationship("Taxon", back_populates="dynamic_properties")

    def __repr__(self):
        return f"<DynamicProperty(key={self.key}, value={self.value})>"
```

```python
DATABASE_URL = "sqlite:///taxon_data.db"  
engine = create_engine(DATABASE_URL, echo=True)

# Create tables in the database
Base.metadata.create_all(engine)

# Create a session
Session = sessionmaker(bind=engine)
session = Session()
```

Data Model for SQL Database

```python
from sqlalchemy import Engine, create_engine
from sqlalchemy.orm import sessionmaker, Session
```

```python
from sqlalchemy import (
    Column, String, Integer, Boolean, Float, ForeignKey, Text, create_engine, JSON
)
from sqlalchemy.orm import relationship, DeclarativeBase, mapped_column
from sqlalchemy.orm import Mapped
```

```python
class Base(DeclarativeBase):
    pass

class Taxon(Base):
    __tablename__ = "taxon"
    
    taxonid = Column(Integer, primary_key=True)
    scientificname = Column(String(255))  # Example column for taxon name
    related_taxons = relationship("RelatedTaxon", back_populates="taxon")

class RelationType(Base):
    __tablename__ = "relation_type"
    
    id = Column(Integer, primary_key=True)
    name = Column(String(100))  # e.g., synonym, parent-child, misapplied name
    
    # Relationship with RelatedTaxon
    related_taxons = relationship("RelatedTaxon", back_populates="relation_type")

class RelatedTaxon(Base):
    __tablename__ = "related_taxon"
    
    id = Column(Integer, primary_key=True)
    taxonid = Column(Integer, ForeignKey("taxon.taxonid"))
    relatednameusageid = Column(Integer, ForeignKey("taxon.taxonid"))  # Refers to another taxon or related name
    relationtype_id = Column(Integer, ForeignKey("relation_type.id"))
    remarks = Column(String(255))  # Any additional notes about the relationship
    
    # Relationships
    taxon = relationship("Taxon", back_populates="related_taxons")
    related_taxon = relationship("Taxon", foreign_keys=[relatednameusageid])
    relation_type = relationship("RelationType", back_populates="related_taxons")

    def __repr__(self):
        return f"<RelatedTaxon(taxonid={self.taxonid}, relatednameusageid={self.relatednameusageid}, relationtype={self.relation_type.name})>"
```

```python
DATABASE_URL = "sqlite:///replacement_name_data.db"  
engine = create_engine(DATABASE_URL, echo=True)

# Create tables in the database
Base.metadata.create_all(engine)

# Create a session
Session = sessionmaker(bind=engine)
session = Session()
```

```python
from sqlalchemy import (
    Column, String, Integer, Boolean, Float, ForeignKey, Text, create_engine
)
from sqlalchemy.orm import relationship, DeclarativeBase, mapped_column
from sqlalchemy.orm import Mapped
```

```python
class Base(DeclarativeBase):
    pass

class Family(Base):
    __tablename__ = "family"
    id: Mapped[int] = mapped_column(primary_key=True, autoincrement=True)
    name: Mapped[str] = mapped_column(unique=True)

    # Relationship with Taxon
    taxons: Mapped[list["Taxon"]] = relationship(
        back_populates="family",
        cascade="all, delete-orphan"
    )

    def __repr__(self) -> str:
        return f"<Family: name={self.name}>"

class Genus(Base):
    __tablename__ = "genus"
    id: Mapped[int] = mapped_column(primary_key=True, autoincrement=True)
    name: Mapped[str] = mapped_column(unique=True)

    # Relationship with Taxon
    taxons: Mapped[list["Taxon"]] = relationship(
        back_populates="genus",
        cascade="all, delete-orphan"
    )

    def __repr__(self) -> str:
        return f"<Genus: name={self.name}>"

class Author(Base):
    __tablename__ = "author"
    id: Mapped[int] = mapped_column(primary_key=True, autoincrement=True)
    name: Mapped[str] = mapped_column(unique=True)

    # Relationship with Taxon
    primary_taxons: Mapped[list["Taxon"]] = relationship(
        back_populates="primary_author",
        primaryjoin="Author.id == Taxon.primary_author_id"
    )
    parenthetical_taxons: Mapped[list["Taxon"]] = relationship(
        back_populates="parenthetical_author",
        primaryjoin="Author.id == Taxon.parenthetical_author_id"
    )

    def __repr__(self) -> str:
        return f"<Author: name={self.name}>"

class Publication(Base):
    __tablename__ = "publication"
    id: Mapped[int] = mapped_column(primary_key=True, autoincrement=True)
    title: Mapped[str] = mapped_column()
    place: Mapped[str] = mapped_column()
    volume_page: Mapped[str] = mapped_column()
    year: Mapped[str] = mapped_column()

    # Relationship with Taxon
    taxons: Mapped[list["Taxon"]] = relationship(
        back_populates="publication",
    )

    def __repr__(self) -> str:
        return f"<Publication: title={self.title}, year={self.year}>"

class GeographicArea(Base):
    __tablename__ = "geographic_area"
    id: Mapped[int] = mapped_column(primary_key=True, autoincrement=True)
    name: Mapped[str] = mapped_column(unique=True)

    # Relationship with Taxon
    taxons: Mapped[list["Taxon"]] = relationship(
        back_populates="geographic_area",
        cascade="all, delete-orphan"
    )

    def __repr__(self) -> str:
        return f"<GeographicArea: name={self.name}>"

class LifeformDescription(Base):
    __tablename__ = "lifeform_description"
    id: Mapped[int] = mapped_column(primary_key=True, autoincrement=True)
    description: Mapped[str] = mapped_column(unique=True)

    # Relationship with Taxon
    taxons: Mapped[list["Taxon"]] = relationship(
        back_populates="lifeform_description",
        cascade="all, delete-orphan"
    )

    def __repr__(self) -> str:
        return f"<LifeformDescription: description={self.description}>"

class ClimateDescription(Base):
    __tablename__ = "climate_description"
    id: Mapped[int] = mapped_column(primary_key=True, autoincrement=True)
    description: Mapped[str] = mapped_column(unique=True)

    # Relationship with Taxon
    taxons: Mapped[list["Taxon"]] = relationship(
        back_populates="climate_description",
        cascade="all, delete-orphan"
    )

    def __repr__(self) -> str:
        return f"<ClimateDescription: description={self.description}>"

class Species(Base):
    __tablename__ = "species"
    id: Mapped[int] = mapped_column(primary_key=True, autoincrement=True)
    name: Mapped[str] = mapped_column(unique=True)

    # Relationship with Taxon
    taxons: Mapped[list["Taxon"]] = relationship(
        back_populates="species",
        cascade="all, delete-orphan"
    )

    def __repr__(self) -> str:
        return f"<Species: name={self.name}>"

class Taxon(Base):
    __tablename__ = "taxon"
    id: Mapped[int] = mapped_column(primary_key=True, autoincrement=True)
    plant_name_id: Mapped[int] = mapped_column(unique=True)
    taxon_name: Mapped[str] = mapped_column(unique=True)
    reviewed: Mapped[bool] = mapped_column()
    taxon_rank: Mapped[str] = mapped_column()
    taxon_status: Mapped[str] = mapped_column()
    accepted_plant_name_id: Mapped[float] = mapped_column()
    basionym_plant_name_id: Mapped[float] = mapped_column()
    replaced_synonym_author: Mapped[str] = mapped_column()
    homotypic_synonym: Mapped[bool] = mapped_column()
    parent_plant_name_id: Mapped[float] = mapped_column()
    powo_id: Mapped[str] = mapped_column()
    hybrid_formula: Mapped[str] = mapped_column()

    # Foreign keys
    family_id: Mapped[int] = mapped_column(ForeignKey("family.id"))
    genus_id: Mapped[int] = mapped_column(ForeignKey("genus.id"))
    primary_author_id: Mapped[int] = mapped_column(ForeignKey("author.id"))
    parenthetical_author_id: Mapped[int] = mapped_column(ForeignKey("author.id"))
    publication_id: Mapped[int] = mapped_column(ForeignKey("publication.id"))
    geographic_area_id: Mapped[int] = mapped_column(ForeignKey("geographic_area.id"))
    lifeform_description_id: Mapped[int] = mapped_column(ForeignKey("lifeform_description.id"))
    climate_description_id: Mapped[int] = mapped_column(ForeignKey("climate_description.id"))
    species_id: Mapped[int] = mapped_column(ForeignKey("species.id"))

    # Relationships
    family: Mapped[Family] = relationship(back_populates="taxons")
    genus: Mapped[Genus] = relationship(back_populates="taxons")
    primary_author: Mapped[Author] = relationship(back_populates="primary_taxons")
    parenthetical_author: Mapped[Author] = relationship(back_populates="parenthetical_taxons")
    publication: Mapped[Publication] = relationship(back_populates="taxons")
    geographic_area: Mapped[GeographicArea] = relationship(back_populates="taxons")
    lifeform_description: Mapped[LifeformDescription] = relationship(back_populates="taxons")
    climate_description: Mapped[ClimateDescription] = relationship(back_populates="taxons")
    species: Mapped[Species] = relationship(back_populates="taxons")

    def __repr__(self) -> str:
        return f"<Taxon: taxon_name={self.taxon_name}, plant_name_id={self.plant_name_id}>"
```

```python
class Base(DeclarativeBase):
    pass

class Family(Base):
    __tablename__ = "family"
    id: Mapped[int] = mapped_column(primary_key=True, autoincrement=True)
    name: Mapped[str] = mapped_column(unique=True)

    # Relationship with Taxon
    taxons: Mapped[list["Taxon"]] = relationship(
        back_populates="family",
        cascade="all, delete-orphan"
    )

    def __repr__(self) -> str:
        return f"<Family: name={self.name}>"

class Genus(Base):
    __tablename__ = "genus"
    id: Mapped[int] = mapped_column(primary_key=True, autoincrement=True)
    name: Mapped[str] = mapped_column(unique=True)

    # Relationship with Taxon
    taxons: Mapped[list["Taxon"]] = relationship(
        back_populates="genus",
        cascade="all, delete-orphan"
    )

    def __repr__(self) -> str:
        return f"<Genus: name={self.name}>"

class Author(Base):
    __tablename__ = "author"
    id: Mapped[int] = mapped_column(primary_key=True, autoincrement=True)
    name: Mapped[str] = mapped_column(unique=True)

    # Relationship with Taxon
    primary_taxons: Mapped[list["Taxon"]] = relationship(
        back_populates="primary_author",
        primaryjoin="Author.id == Taxon.primary_author_id"
    )
    parenthetical_taxons: Mapped[list["Taxon"]] = relationship(
        back_populates="parenthetical_author",
        primaryjoin="Author.id == Taxon.parenthetical_author_id"
    )

    def __repr__(self) -> str:
        return f"<Author: name={self.name}>"

class Publication(Base):
    __tablename__ = "publication"
    id: Mapped[int] = mapped_column(primary_key=True, autoincrement=True)
    title: Mapped[str] = mapped_column()
    place: Mapped[str] = mapped_column()
    volume_page: Mapped[str] = mapped_column()
    year: Mapped[str] = mapped_column()

    # Relationship with Taxon
    taxons: Mapped[list["Taxon"]] = relationship(
        back_populates="publication",
    )

    def __repr__(self) -> str:
        return f"<Publication: title={self.title}, year={self.year}>"


class LifeformDescription(Base):
    __tablename__ = "lifeform_description"
    id: Mapped[int] = mapped_column(primary_key=True, autoincrement=True)
    description: Mapped[str] = mapped_column(unique=True)

    # Relationship with Taxon
    taxons: Mapped[list["Taxon"]] = relationship(
        back_populates="lifeform_description",
        cascade="all, delete-orphan"
    )

    def __repr__(self) -> str:
        return f"<LifeformDescription: description={self.description}>"

class ClimateDescription(Base):
    __tablename__ = "climate_description"
    id: Mapped[int] = mapped_column(primary_key=True, autoincrement=True)
    description: Mapped[str] = mapped_column(unique=True)

    # Relationship with Taxon
    taxons: Mapped[list["Taxon"]] = relationship(
        back_populates="climate_description",
        cascade="all, delete-orphan"
    )

    def __repr__(self) -> str:
        return f"<ClimateDescription: description={self.description}>"

class Species(Base):
    __tablename__ = "species"
    id: Mapped[int] = mapped_column(primary_key=True, autoincrement=True)
    name: Mapped[str] = mapped_column(unique=True)

    # Relationship with Taxon
    taxons: Mapped[list["Taxon"]] = relationship(
        back_populates="species",
        cascade="all, delete-orphan"
    )

    def __repr__(self) -> str:
        return f"<Species: name={self.name}>"


class InfraspeciesRank(Base):
    __tablename__ = "infraspecies_rank"
    id: Mapped[int] = mapped_column(primary_key=True, autoincrement=True)
    rank: Mapped[str] = mapped_column(unique=True)

    taxons: Mapped[list["Taxon"]] = relationship(
        back_populates="infraspecies_rank",
        cascade="all, delete-orphan"
    )

    def __repr__(self) -> str:
        return f"<InfraspeciesRank: rank={self.rank}>"

class Infraspecies(Base):
    __tablename__ = "infraspecies"
    id: Mapped[int] = mapped_column(primary_key=True, autoincrement=True)
    name: Mapped[str] = mapped_column(unique=True)

    taxons: Mapped[list["Taxon"]] = relationship(
        back_populates="infraspecies",
        cascade="all, delete-orphan"
    )

    def __repr__(self) -> str:
        return f"<Infraspecies: name={self.name}>"

class TaxonStatus(Base):
    __tablename__ = "taxon_status"
    id: Mapped[int] = mapped_column(primary_key=True, autoincrement=True)
    status: Mapped[str] = mapped_column(unique=True)

    taxons: Mapped[list["Taxon"]] = relationship(
        back_populates="taxon_status",
        cascade="all, delete-orphan"
    )

    def __repr__(self) -> str:
        return f"<TaxonStatus: status={self.status}>"

class Taxon(Base):
    __tablename__ = "taxon"

    # Columns
    id: Mapped[int] = mapped_column(primary_key=True, autoincrement=True)
    plant_name_id: Mapped[int] = mapped_column(unique=True)
    taxon_name: Mapped[str] = mapped_column(unique=True)
    reviewed: Mapped[bool] = mapped_column()
    taxon_rank: Mapped[str] = mapped_column()
    taxon_status: Mapped[str] = mapped_column()
    accepted_plant_name_id: Mapped[float] = mapped_column()
    basionym_plant_name_id: Mapped[float] = mapped_column()
    replaced_synonym_author: Mapped[str] = mapped_column()
    homotypic_synonym: Mapped[bool] = mapped_column()
    parent_plant_name_id: Mapped[float] = mapped_column()
    powo_id: Mapped[str] = mapped_column()
    hybrid_formula: Mapped[str] = mapped_column()

        #columns from distribution.csv
    introduced: Mapped[bool] = mapped_column()
    extinct: Mapped[bool] = mapped_column()
    location_doubtful: Mapped[bool] = mapped_column()

    # Foreign keys
    family_id: Mapped[int] = mapped_column(ForeignKey("family.id"))
    genus_id: Mapped[int] = mapped_column(ForeignKey("genus.id"))
    primary_author_id: Mapped[int] = mapped_column(ForeignKey("author.id"))
    parenthetical_author_id: Mapped[int] = mapped_column(ForeignKey("author.id"))
    publication_id: Mapped[int] = mapped_column(ForeignKey("publication.id"))
    geographic_area_id: Mapped[int] = mapped_column(ForeignKey("geographic_area.id"))
    lifeform_description_id: Mapped[int] = mapped_column(ForeignKey("lifeform_description.id"))
    climate_description_id: Mapped[int] = mapped_column(ForeignKey("climate_description.id"))
    species_id: Mapped[int] = mapped_column(ForeignKey("species.id"))
    infraspecies_rank_id: Mapped[int] = mapped_column(ForeignKey("infraspecies_rank.id"))
    infraspecies_id: Mapped[int] = mapped_column(ForeignKey("infraspecies.id"))
    taxon_status_id: Mapped[int] = mapped_column(ForeignKey("taxon_status.id"))

    # Relationships
    family: Mapped[Family] = relationship(back_populates="taxons")
    genus: Mapped[Genus] = relationship(back_populates="taxons")
    primary_author: Mapped[Author] = relationship(back_populates="primary_taxons")
    parenthetical_author: Mapped[Author] = relationship(back_populates="parenthetical_taxons")
    publication: Mapped[Publication] = relationship(back_populates="taxons")
    geographic_area: Mapped[GeographicArea] = relationship(back_populates="taxons")
    lifeform_description: Mapped[LifeformDescription] = relationship(back_populates="taxons")
    climate_description: Mapped[ClimateDescription] = relationship(back_populates="taxons")
    species: Mapped[Species] = relationship(back_populates="taxons")
    infraspecies_rank: Mapped["InfraspeciesRank"] = relationship(back_populates="taxons")
    infraspecies: Mapped["Infraspecies"] = relationship(back_populates="taxons")
    taxon_status: Mapped["TaxonStatus"] = relationship(back_populates="taxons")

    def __repr__(self) -> str:
        return f"<Taxon: taxon_name={self.taxon_name}, plant_name_id={self.plant_name_id}>"

class Continent(Base):
    __tablename__ = "continent"
    id: Mapped[int] = mapped_column(primary_key=True, autoincrement=True)
    code: Mapped[int] = mapped_column(unique=True)  # continent_code_l1
    name: Mapped[str] = mapped_column(unique=True)  # continent

    # Relationship with GeographicArea
    geographic_area: Mapped[list["GeographicArea"]] = relationship(
        back_populates="continent",
        cascade="all, delete-orphan"
    )

    def __repr__(self) -> str:
        return f"<Continent: name={self.name}, code={self.code}>"
    
class Region(Base):
    __tablename__ = "region"
    id: Mapped[int] = mapped_column(primary_key=True, autoincrement=True)
    code: Mapped[float] = mapped_column(unique=True)  # region_code_l2
    name: Mapped[str] = mapped_column(unique=True)    # region

    # Relationship with WCVPDistribution
    geographic_area: Mapped[list["GeographicArea"]] = relationship(
        back_populates="region",
        cascade="all, delete-orphan"
    )

    def __repr__(self) -> str:
        return f"<Region: name={self.name}, code={self.code}>"
    
class Area(Base):
    __tablename__ = "area"
    id: Mapped[int] = mapped_column(primary_key=True, autoincrement=True)
    code: Mapped[str] = mapped_column(unique=True, nullable=True)  # area_code_l3
    name: Mapped[str] = mapped_column(unique=True, nullable=True)  # area

    # Relationship with WCVPDistribution
    geographic_area: Mapped[list["GeographicArea"]] = relationship(
        back_populates="area",
        cascade="all, delete-orphan"
    )

    def __repr__(self) -> str:
        return f"<Area: name={self.name}, code={self.code}>"
    
class GeographicArea(Base):
    __tablename__ = "geographic_area"
    id: Mapped[int] = mapped_column(primary_key=True, autoincrement=True)
    name: Mapped[str] = mapped_column(unique=True)

    # Foreign keys
    area_id: Mapped[int] = mapped_column(ForeignKey("area.id"))
    region_id: Mapped[int] = mapped_column(ForeignKey("region.id"))
    continent_id: Mapped[int] = mapped_column(ForeignKey("continent.id")) 

    # Relationship with Taxon
    taxons: Mapped[list["Taxon"]] = relationship(
        back_populates="geographic_area",
        cascade="all, delete-orphan"
    )
    # Relationship with Area
    area: Mapped[Area] = relationship(back_populates="geographic_area")
    # Relationship with Region
    region: Mapped[Region] = relationship(back_populates="geographic_area")
    # Relationship with Continent
    continent: Mapped[Continent] = relationship(back_populates="geographic_area")

    def __repr__(self) -> str:
        return f"<GeographicArea: name={self.name}>"
```
